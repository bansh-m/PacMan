```python
import pandas
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
f = open('stat.csv')
dataFrame_in = pandas.read_csv(f)
pandas.options.display.float_format = '{:,.0f}'.format
count = 5
dataFrame_in
```

```python
dataFrame = dataFrame_in.copy()
algoList = dataFrame['Algorithm'].tolist()
statusList = dataFrame['Status'].tolist()
```

```python
labelEncoderAlgo = LabelEncoder()
labelEncoderAlgo.fit(algoList)
labelsAlgo = labelEncoderAlgo.transform(algoList)
dataFrame['Algorithm']=pandas.Series(labelsAlgo)

labelEncoderStatus = LabelEncoder()
labelEncoderStatus.fit(statusList)
labelStatus= labelEncoderStatus.transform(statusList)
dataFrame['Status']=pandas.Series(labelStatus)

dataFrame
```

```python
x = np.array(dataFrame[['CrossroadsCount']])
y = np.array(dataFrame[['Status', 'Score', 'Time', 'Algorithm']])

newCrossroadsValues = [[70], [80], [90], [100], [110], [120]]
model = LinearRegression()
model.fit(x, y)
prediction = pandas.DataFrame(model.predict(newCrossroadsValues), columns=['Status', 'Score', 'Time', 'Algorithm'])
prediction['CrossroadsCount'] = newCrossroadsValues

prediction
```

```python
crossroadsValues = list()
for crossroadListValue in prediction['CrossroadsCount']:
    crossroadsValues.append(crossroadListValue[0])
prediction['CrossroadsCount'] = crossroadsValues
convert_dict = {'Status': int,
                'Score': int,
                'Time': float,
                'Algorithm': int,
                'CrossroadsCount': int
                }
prediction = prediction.astype(convert_dict)

prediction
```

```python
result = dataFrame.append(prediction)

def denormalize_status(value):
       if value in [0, '0']:
           return 'lose'
       else:
           return 'win'

def denormalize_algorithm(value):
   if value in [0, '0']:
       return 'Minimax'
   else:
       return 'Expectimax'

statusValues = result['Status']
AlgorithmValues = result['Algorithm']
newStatusValues = list()
newAlgValues = list()
for status in statusValues:
    newStatusValues.append(denormalize_status(status))
for algValue in AlgorithmValues:
    newAlgValues.append(denormalize_algorithm(algValue))
result['Status'] = newStatusValues
result['Algorithm'] = newAlgValues

result.to_csv('out.csv')

result
```
